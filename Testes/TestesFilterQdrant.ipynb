```python
from sentence_transformers import SentenceTransformer
from qdrant_client import QdrantClient, models
```

```python
model_name = "Qwen/Qwen3-Embedding-0.6B"
encoder = SentenceTransformer(model_name, device='cpu')
```

```python
client = QdrantClient(path="../metadados/VectorStore")
```

```python
query="doenças endemicas" 

hits = client.query_points(
    collection_name="Recurso_metadados",
    query=encoder.encode(query).tolist(),
    limit=20
).points

for hit in hits: # Resultados não são ruins, mas podem ser bem melhores
    print(hit.score)
    print(hit.payload.get('titulo'))
    print(hit.payload.get('descricao'))
    print(hit.payload.get('formato'))
    print()
```

```python
query = "Doenças endêmicas"

tag_hits = client.query_points(
    collection_name="Tags_metadados",
    query=encoder.encode(query).tolist(),
    limit=20
).points

for hit in tag_hits: # Retorna resultads mais condizentes com doenças endemicas, pode ser usado na pesquisa de datasets
    print(list(hit.payload.keys()))
```
